```python
#IMPORTING ALL THE ESSENTIAL LIBRARIES FOR OUR PROJECT

import numpy as np
import matplotlib as plt
import seaborn as sb
import pandas as pd
%matplotlib inline
```

Loading the Dataset

```python
colunms = ['Sepal lenght','Sepal width','Petal lenght','Petal width','Class_labels']
#load the data
df = pd.read_csv('/home/Iris.csv', names=colunms)
df = df.replace(r'^\s*$', np.nan, regex=True)
df.head()
```

Visualization of our Data

```python
df.describe()
```

```python
#visualiza the whole dataset
sb.pairplot(df, hue ='Class_labels')
```

Seperating Input Columns and the Output Columns

```python
data = df.values

X = data[:,0:4]
Y = data [:,4]
print(Y)
```

Splitting the Data into Training and Testing

```python
#Split the data to train and test the dataset
from sklearn.model_selection import train_test_split
X_train , X_test ,y_train, y_test = train_test_split(X,Y, test_size = 0.2)
print(X_test)
```

MODEL1 : SUPPORT VECTOR MACHINE LEARNING ALGORITHM

```python
#support vector machine learning algorithm
from sklearn.svm import SVC

model_svc = SVC()
model_svc.fit(X_train , y_train)
```

```python
prediction1 = model_svc.predict(X_test)
#calculate the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, prediction1)*100)

for i in range(len(prediction1)):
 print(y_test[i],prediction1[i])
```
